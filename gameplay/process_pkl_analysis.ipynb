```python
import pickle
import pandas as pd
```

```python
data = pickle.load(open("../Engagement/FaceEngageDataset/u1/u1_00001.pkl", "rb"))
```

```python
EMOTIONS = ["Neutral", "Happy", "Sad", "Surprise", "Fear", "Disgust", "Anger", "Contempt"]
```

```python
last_frame = max(data["Frame Index"])
df = pd.DataFrame(index=pd.RangeIndex(last_frame+1), columns=["valence", "arousal"]+EMOTIONS)
idx = data["Frame Index"]
df["valence"].loc[idx] = data["Valence"]
df["arousal"].loc[idx] = data["Arousal"]
for i, e in enumerate(EMOTIONS):
    df[e].loc[idx] = data["Emotions"][:, i]
df = df.reset_index().rename(columns={"index":"frame"})
df
```
